#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# calculate % change per london Borough
properties_per_Borough_first = properties_8.groupby('London Borough').first()
print(properties_per_Borough_first)
properties_per_Borough_last = properties_8.groupby('London Borough').last()
print(properties_per_Borough_last)

properties_per_Borough = pd.merge(properties_per_Borough_first, properties_per_Borough_last, on=['London Borough', 'ID'], how='outer')
print(properties_per_Borough)
print(properties_per_Borough.columns)

properties_per_Borough['% Change'] = properties_per_Borough.apply(lambda x : (x['Average Price_y']-x['Average Price_x'])/x['Average Price_x']*100, axis=1)

# need axis=1 in apply otherwise will get key error
# refer to 'https://stackoverflow.com/questions/39960728/pandas-apply-key-error'

print(properties_per_Borough)
print(properties_per_Borough.index)
```

```python
# visulize % chnage per london Borough
properties_per_Borough.iloc[:, 5].plot(kind='bar')

plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# define the function create_price_ratio, for properties_9 as input
def create_price_ratio(dfg, London_Borough):
    dfg1 = dfg[dfg['London Borough']==London_Borough]
    ratio = dfg1.loc['2018', 'Average Price'].mean()/dfg1.loc['1998', 'Average Price'].mean() # this is just one number!
    return ratio
        
London_Borough = properties_9['London Borough'].unique()
# dfg = pd.DataFrame(columns=['London Borough','2018/1998 Ratio'])
Borough = []
Ratio = []

for i in London_Borough:
    Borough.append(i) 
    Ratio.append(create_price_ratio(properties_9, i))

dfg = pd.DataFrame({'London Borough': Borough ,'2018/1998 Ratio': Ratio})
print(dfg) 
```

```python
# visualize the dfg (2018/1998 Ratio per London Borough) in decending order
dfg_ordered = dfg.sort_values('2018/1998 Ratio', ascending=False)
dfg_ordered = dfg_ordered.reset_index(drop=True)
print(dfg_ordered)

dfg_ordered.plot(x='London Borough', y='2018/1998 Ratio', kind='bar')

plt.show()
print(dfg_ordered.shape)
```

```python
# since the results were different from % Change calculated earlier using wider time range (1995-2020), calculate the 2020/1995 ratio to see if they will match

# define the function create_price_ratio, for properties_9 as input
def create_price_ratio1(dfg, London_Borough):
    dfg1 = dfg[dfg['London Borough']==London_Borough]
    ratio = dfg1.loc['2020', 'Average Price'].mean()/dfg1.loc['1995', 'Average Price'].mean() # this is just one number!
    return ratio
        
London_Borough = properties_9['London Borough'].unique()

Borough1 = []
Ratio1 = []

for i in London_Borough:
    Borough1.append(i) 
    Ratio1.append(create_price_ratio1(properties_9, i))

dfg1 = pd.DataFrame({'London Borough': Borough1 ,'2020/1995 Ratio': Ratio1})
print(dfg1) 
```

```python
# visualize the dfg (2018/1998 Ratio per London Borough) in decending order
dfg1_ordered = dfg1.sort_values('2020/1995 Ratio', ascending=False)
dfg1_ordered = dfg1_ordered.reset_index(drop=True)
print(dfg1_ordered)

dfg1_ordered.plot(x='London Borough', y='2020/1995 Ratio', kind='bar')

plt.show()
print(dfg1_ordered.shape)
```
